● Print top 5 rows and column names

```python
import pandas as pd
url="https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv "
df=pd.read_csv(url)

print("Columns",df.columns.tolist())
print(df.head())
```

● Convert time to datetime

```python
df['time'] = pd.to_datetime(df['time'])
df['time'].head()
```

● Drop records with missing latitude, longitude, mag

---

```python
df = df.dropna(subset=['latitude','longitude','mag'])
df
```

```python
df.isna().sum()
```

 Filter only earthquakes with magnitude >= 4.0

```python
df_earthquake = df[df['mag'] >= 4.0]
df_earthquake
```

 Add a column day_of_week from time

```python
df['column day_of_week']=df['time'].dt.day_name()
df
```

● Create a column severity_level based on magnitude:

○ <4.0: "Low"

○ 4.0 - 6.0: "Moderate"

○ 6.0+: "High"

```python
def severity_level(mag):
  if mag < 4.0:
    return 'Low'
  elif 4.0 <= mag <= 6.0:
    return 'Moderate'
  else:
    return 'High'
```

```python
df['serenity_level'] = df['mag'].apply(severity_level)
df['serenity_level']
```

```python
df.info()
```

Count number of earthquakes per place

```python
place_counts = df['place'].value_counts()
print(place_counts.head(10))
```

● Compute average magnitude and max depth per day

```python
daily_summary = df_earthquake.groupby(df_earthquake['time'].dt.date).agg(
    avg_magnitude=('mag', 'mean'),
    max_depth=('depth', 'max')
).reset_index()
print(daily_summary)
```

```python
# Save:
# ○ Cleaned dataset to cleaned_earthquakes.csv
# ○ Summary data to earthquake_summary.csv
```

```python
df_earthquake.to_csv("cleaned_earthquakes.csv", index=False)
daily_summary.to_csv("earthquake_summary.csv", index=False)
```

Store into SQLite table

```python
import sqlite3

conn = sqlite3.connect("earthquakes.db")
df_earthquake.to_sql("earthquakes", conn, if_exists="replace", index=False)
daily_summary.to_sql("summary", conn, if_exists="replace", index=False)
conn.close()
```
