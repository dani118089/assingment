libraries


```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
#buildng logistics regression model()
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score
from sklearn.neighbors import KNeighborsClassifier  #knn
from sklearn.naive_bayes import GaussianNB #naive bayes
from sklearn.svm import SVC #svm
from sklearn import metrics
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier#decision tree
from sklearn.ensemble import RandomForestClassifier  #for random fores
```

```python

filepath='/content/drive/MyDrive/DSA/iris.csv'
df_iris=pd.read_csv(filepath)
df_iris
```

```python
encoder= LabelEncoder()
df_iris['species']=encoder.fit_transform(df_iris['species'])
df_iris.head(3)
```

midel building

```python
#split data into train and test
from sklearn.model_selection import train_test_split

x=df_iris.drop('species', axis=1)
y=df_iris['species']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python

log_reg_model = LogisticRegression()
log_reg_model.fit(x_train, y_train) #training the model
y_pred = log_reg_model.predict(x_test) #predicting the test data
#evaluate the model
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred,average='weighted')
recall = recall_score(y_test, y_pred,average='weighted')

print("Accuracy:", accuracy)
print("precision", precision)
print("recall", recall)
```

```python
#prediction using threhold
y_pred_threshold = log_reg_model.predict_proba(x_test)[:,1]
y_pred_threshold=(y_pred_threshold>0.33).astype(int)
#evaluate the model
accuracy_thresh = accuracy_score(y_test,y_pred_threshold)
precision_thresh = precision_score(y_test,y_pred_threshold,average='weighted')
recall_thresh = recall_score(y_test,y_pred_threshold,average='weighted')

print('threshold value used =0.333 ')
print('accuracy score = ',accuracy_thresh)
print('precision score = ',precision_thresh)
print('recall score = ',recall_thresh)
```

K-NN    (k nearest neigbours)

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)#training the model
y_pred=knn.predict(x_test)#prediction
#evaluate the model
accuracy = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred,average='weighted')
recall = recall_score(y_test,y_pred,average='weighted')

print("Accuracy : ",accuracy)
print("Precision : ",precision)
print("Recall : ",recall)
```

naive bayers classification
using naive bayers

```python
nb_model = GaussianNB()
nb_model.fit (x_train,y_train) #train the model
y_pred_nb = nb_model.predict(x_test) #predict
#evalution
accuracy_nb = accuracy_score(y_test,y_pred_nb)
precision_nb= precision_score(y_test,y_pred_nb,average='weighted')
recall_nb = recall_score(y_test,y_pred_nb,average='weighted')

print("Accuracy : ",accuracy_nb)
print("Precision : ",precision_nb)
print("Recall : ",recall_nb)
```

SVM classifier

```python
svm_model=SVC()
svm_model.fit(x_train,y_train)#training the model
y_pred_svm=svm_model.predict(x_test)#prediction
#Evaluation
accuracy_svm=accuracy_score(y_test,y_pred_svm)
precision_svm=precision_score(y_test,y_pred_svm,average='weighted')
recall_svm=recall_score(y_test,y_pred_svm,average='weighted')
confusion_matrix_svc = confusion_matrix(y_test, y_pred_svm)#confusion martix

print("Accuracy : ",accuracy_svm)
print("Precision : ",precision_svm)
print("Recall : ",recall_svm)
print("confusion matrix :  ")
print(confusion_matrix_svc)
```

Desicion tree and random tree


```python
desicion_model=DecisionTreeClassifier()
desicion_model.fit(x_train,y_train)
y_pred_desicion=desicion_model.predict(x_test)
#evaluation
accuracy_desicion=accuracy_score(y_test,y_pred_desicion)
precision_desicion=precision_score(y_test,y_pred_desicion,average='weighted')
recall_desicion=recall_score(y_test,y_pred_desicion,average='weighted')
confusion_matrix_svc = confusion_matrix(y_test, y_pred_desicion)#confusion martix


print("accuracy : ",accuracy_desicion)
print("precision : ",precision_desicion)
print("recall : ",recall_desicion)
print("confusion matrix :  ")
print(confusion_matrix_svc)
```

```python
random_model=RandomForestClassifier(criterion="gini")#"gini", "entropy", "log_loss"
random_model.fit(x_train,y_train)
y_pred_random=random_model.predict(x_test)
#evaluation
accuracy_random=accuracy_score(y_test,y_pred_random)
precision_random=precision_score(y_test,y_pred_random,average='weighted')
recall_random=recall_score(y_test,y_pred_random,average='weighted')
confusion_matrix_svc = confusion_matrix(y_test, y_pred_random)#confusion martix


print("accuracy : ",accuracy_random)
print("precision : ",precision_random)
print("recall : ",recall_random)
print("confusion matrix :  ")
print(confusion_matrix_svc)
```
